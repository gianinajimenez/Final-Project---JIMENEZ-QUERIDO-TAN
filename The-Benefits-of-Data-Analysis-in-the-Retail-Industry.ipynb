## 1. Maintaining Performance Across Wide Scopes

It is almost all businessmen’s goal to expand their business and to create an empire. However, a risk that comes with expansion is the loss of control due to decentralization. With that, the challenge for businessmen is to monitor each branch in such a way that allows them to maintain quality performance across branches. Data Analysis provides the means for them to do. Through Data Analysis, data can be grouped, packaged and presented in a way that leads to analysis extraction. 

Given the supermarket sales dataset, the tables and graphs below were constructed. The <b>Net Profit per Branch</b> provides an overview of the branches’ performance within the given time span. Considering this, it can be inferred that the overall sales of the branches are almost equal, indicating that none of the branches are underperforming. Hence, there is no need to give more focus on one branch over the other as a whole. 

```python
# Net Profit per Branch 
npb = df.loc[:,["Branch","Net Profit"]].groupby("Branch",as_index=False).sum()

fig, ax = plt.subplots(1)

ax.set_title("Net Profit per Branch")
ax.set_xlabel("Branch")
ax.set_ylabel("Profit")
ax.bar(height=npb["Net Profit"],x=npb["Branch"],color="lightblue")
```

However, looking at the dataset in a different perspective might offer more substantial insights. To elaborate, the <b>Net Profit per Branch Over Time</b> provides a more detailed account of the sales per branch, which highlights when each branch is doing better over the time period. It can be inferred from the line plot below that branches B and C have high sales in the months of January and March, however they decrease in the month of February. Branch A may not make as much profit as B and C, but it's more consistent and stable with a smaller amount of peaks. The highest peak of Branch B is on 01/26/2019, and from here, the business could identify the factors that could have led to that high number. It could be due to good work practices, holidays, discounts, and more. On days with low sales, managers can identify what the branch did wrong or what they can do better on days with bad conditions.

With this, we see that data analysis can be used as a tool for visualizing patterns and trends across different branches, and it helps managers identify significant problems and occurences and make practical decisions based on real data. 


```python
# Net Profit per Branch Over Time 
df1=df.groupby(["Date","Month","Branch"],as_index=False).sum()
df1=df1.loc[:,["Date","Month","Branch","Net Profit"]]
plt.figure(figsize=(20, 5))
ax = sns.lineplot(x="Date", y="Net Profit", hue="Branch",data=df1,linewidth=2)
ax.set_title("Net Profit per Branch Over Time",fontsize=20)
ax.set(xticks=df1.Date[3::9])
for item in ax.get_xticklabels():
        item.set_rotation(90)
```

## 3. Improving & Personalizing Service

As much as the retail industry is about selling its products, the value of the service aspect of the industry is slowly coming into light due to the gradually increasing competitiveness of the market. As a result, customers are becoming increasingly harder to please. On top of this, companies not only have to compete with other physical stores, but also brands operating online. According to a [source](https://smallbiztrends.com/2019/04/things-customers-want-retail-stores.html), "Customers want in-person service they can’t get online. This is crucial to differentiating your store from the one-dimensional online shopping experience." With that, good service has become the bare minimum and providing personalized service has become a necessity. Business owners have to give in to the increasing demands of its consumers because [studies](https://www.widerfunnel.com/customer-experience/) show that it is 6-7 times more expensive to attract a new customer than it is to retain an existing one. Thus, business owners must make it a priority to provide nothing but the best service to its customers and Data Analysis serves as a tool for business owners to do just that.

According to a [study](https://www.supermarketnews.com/retail-financial/customers-waiting-hardest-part), the waiting is the hardest part for customers.  Through Data Analysis, customer service can be improved by shortening check-out waiting time. The figures below present the <b>Number of Sales per Payment Type</b>. Based on the figures, it can be deduced that the most popular payment type is through e-wallet, with cash falling closely behind. From this, one recommendation could be to increase the number of counters dedicated solely for e-wallet or cash customers. This would lead to faster transactions and more satisfied customers.

Furthermore, Data Analysis can also be used to make customer service more personalized. The figures below show the <b>Number of Members per Gender</b> and the <b>Total Amount Spent per Gender</b>. The graph of the former underlines that there are more females than males who are members of the supermarket. With this knowledge, the membership service can be personalized in such a way that would cater to more women. On the other hand, the graph of the latter highlights that females generally spend more in the store than males. As a response to this, the company can deliver more relevant products or a wider range of choices for women shoppers in order to service them better. By using Data Analysis to better meet consumer needs, consumers will be more loyal to the brand, which will then be reflected in the increased sales of the business. 

```python
nspt = df.loc[:,["Invoice ID","Payment"]].groupby("Payment",as_index=False).count()

fig, ax = plt.subplots(1)

ax.set_title("Number of Sales per Payment Type")
ax.set_xlabel("Payment Type")
ax.set_ylabel("Number of Sales")
ax.bar(height=nspt["Invoice ID"],x=nspt["Payment"],color="lightblue")

nmg = df.loc[df.loc[:,"Customer Type"]=="Member",["Customer Type","Gender"]].groupby("Gender",as_index=False).count()

fig, ax = plt.subplots(1)

ax.set_title("Number of Members per Gender")
ax.set_xlabel("Gender")
ax.set_ylabel("Number of Member")
ax.bar(height=nmg["Customer Type"],x=nmg["Gender"],color="lightblue")

tasg = df.loc[:,["Total","Gender"]].groupby("Gender",as_index=False).sum()

fig, ax = plt.subplots(1)

ax.set_title("Total Amount Spent per Gender")
ax.set_xlabel("Gender")
ax.set_ylabel("Total Amount Spent")
ax.bar(height=tasg["Total"],x=tasg["Gender"],color="lightblue")
```

## 5. Optimizing Employee Shift Schedules

Unbeknownst to many, workforce scheduling greatly impacts a retail store’s sales and customer service. According to a [survey](https://www.widerfunnel.com/customer-experience/) , 66% of consumers switch brands because of poor service. As the market becomes more and more condensed, consumers are given more options that make it convenient for them to easily switch brands. Considering this, hiring the best staff and ensuring staff availability has become increasingly more important. However, business owners should optimize in hiring the right amount of workers in certain schedules. A surplus in workers during work hours with less foot traffic and less sales would lead to profit loss. Similarly, a lack of workers during peak hours would lead to the inability to provide quality service for customers at the given time period, leading to customer dissatisfaction and eventually to profit loss. 

Given this, Data Analysis can be a helpful tool that can be used to optimize employee shift schedules as seen in the figures below. By displaying the <b>Number of Sales per Hour</b>, the business owner can identify which hours have the most number of customers and which hours have lesser than average customers. In order to optimize shift schedules, the owner can have different amounts of workers in different timestamps during the day, where more workers are assigned during busy timestamps and less in not-so busy hours so as to not waste resources. To be more specific, in the dataset given, more employees should be hired from 10-11 am, 1-2 pm, 3-4 pm and 7-8 pm.

```python
ten = df.loc[df["Time"].str.match("10"),"Invoice ID"].count()
eleven = df.loc[df["Time"].str.match("11"),"Invoice ID"].count()
twelve = df.loc[df["Time"].str.match("12"),"Invoice ID"].count()
thirteen = df.loc[df["Time"].str.match("13"),"Invoice ID"].count()
fourteen = df.loc[df["Time"].str.match("14"),"Invoice ID"].count()
fifteen = df.loc[df["Time"].str.match("15"),"Invoice ID"].count()
sixteen = df.loc[df["Time"].str.match("16"),"Invoice ID"].count()
seventeen = df.loc[df["Time"].str.match("17"),"Invoice ID"].count()
eighteen = df.loc[df["Time"].str.match("18"),"Invoice ID"].count()
nineteen = df.loc[df["Time"].str.match("19"),"Invoice ID"].count()
twenty = df.loc[df["Time"].str.match("20"),"Invoice ID"].count()
nsh = pd.DataFrame({"Hour":["10 am - 11 am","11 am - 12 pm","12 pm - 1 pm", "1 pm - 2 pm", "2 pm - 3 pm", "3 pm - 4 pm", "4 pm - 5 pm", "5 pm - 6 pm", "6 pm - 7 pm","7 pm - 8 pm", "8 pm - 9 pm"],"Count":[ten,eleven,twelve,thirteen,fourteen,fifteen,sixteen,seventeen,eighteen,nineteen,twenty]})

import matplotlib.pyplot as plt

fig, ax = plt.subplots(1)

ax.set_title("Number of Sales per Hour")
ax.set_xlabel("Hour")
ax.set_ylabel("Number of Sales")
ax.bar(height=nsh["Count"],x=nsh["Hour"],color="lightblue")
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
```

## 6. Deriving Richer Insights Through Cross Referencing

Data Analysis can provide businessmen with richer insights by studying the relationships between different correlations of a dataset. Oftentimes, it is not enough to analyze only one relationship, as it limits the insight that one is able to derive from a dataset. The analysis of the figures below will attest to this.

Firstly, the figures below display the relationship between the <b>Number of Sales per Month</b> as compared to the <b>Net Profit per Month</b>. The graph of the Number of Sales per Month suggests that the order of most sales per month is as follows: January, March and February. Similarly, the graph of the Net Profit per Month shows that the order of the most net profit per month is also the same as the results from the previous graph. As such, by analyzing both figures simultaneously, it can be inferred that there is not much disparity in how much each customer spends, meaning there is little to no outliers in the dataset. From this, it is implied that the majority of the customers of that specific supermarket has the same financial standing, or is at least able to spend at around the average total which is P749.30. By identifying the customer profile, the supermarket will be able to better meet the demands of their target market, which can possibly improve their overall profitability and customer satisfaction. 


```python
nsm = df.loc[:,["Invoice ID","Month"]].groupby("Month",as_index=False).count()

fig, ax = plt.subplots(1)

ax.set_title("Number of Sales per Month")
ax.set_xlabel("Month")
ax.set_ylabel("Number of Sales")
ax.bar(height=nsm["Invoice ID"],x=nsm["Month"],color="lightblue")

npm = df.loc[:,["Net Profit","Month"]].groupby("Month",as_index=False).sum()

fig, ax = plt.subplots(1)

ax.set_title("Net Profit per Month")
ax.set_xlabel("Month")
ax.set_ylabel("Net Profit")
ax.bar(height=npm["Net Profit"],x=npm["Month"],color="lightblue")

```

Secondly, the figures below display the relationship between the <b>Number of Sales per Payment Type</b> and the <b>Net Profit per Payment Type</b>. If the Number of Sales per Payment Type is analyzed on its own, it gives the impression that the e-wallet type of payment should be given most focus on because it garners the most amount of sales. However, when the Net Profit per Payment Type is also taken into consideration, it can be seen that it is the cash form of payment, not the e-wallet payment, that garners the most net profit. Thus, this reminds the business owner that cash as a form of payment type should equally be prioritized. 

```python
nspt = df.loc[:,["Invoice ID","Payment"]].groupby("Payment",as_index=False).count()

fig, ax = plt.subplots(1)

ax.set_title("Number of Sales per Payment Type")
ax.set_xlabel("Payment Type")
ax.set_ylabel("Number of Sales")
ax.bar(height=nspt["Invoice ID"],x=nspt["Payment"],color="lightblue")

nppt = df.loc[:,["Net Profit","Payment"]].groupby("Payment",as_index=False).sum()

fig, ax = plt.subplots(1)

ax.set_title("Net Profit per Payment Type")
ax.set_xlabel("Payment Type")
ax.set_ylabel("Net Profit")
ax.bar(height=nppt["Net Profit"],x=nppt["Payment"],color="lightblue")
```
